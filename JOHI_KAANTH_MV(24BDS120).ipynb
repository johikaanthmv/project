```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/dataset_phishing.csv.zip")
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

df['num_letters_url'] = df['url'].apply(lambda x: sum(c.isalpha() for c in str(x)))

df['ratio_letters_url'] = df.apply(lambda row: row['num_letters_url'] / row['length_url'] if row['length_url'] > 0 else 0, axis=1)

plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
plt.plot(df['num_letters_url'][:500], df['ratio_letters_url'][:500])
plt.title("Line Plot: Number of Letters in URL vs Ratio of Letters in URL")
plt.xlabel("Number of Letters in URL")
plt.ylabel("Ratio of Letters in URL")


plt.subplot(2, 2, 2)
pie_values = df['ratio_letters_url'][:5].abs()
plt.pie(pie_values, labels=[f"Sample {i+1}" for i in range(5)])
plt.title("Pie Chart of First 5 Letter Ratios in URL")

plt.subplot(2, 2, 3)
plt.bar(df['num_letters_url'][:10], df['ratio_letters_url'][:10])
plt.title("Bar Chart (First 10 Samples)")
plt.xlabel("Number of Letters in URL")
plt.ylabel("Ratio of Letters in URL")


plt.subplot(2, 2, 4)
plt.scatter(df['num_letters_url'][:500], df['ratio_letters_url'][:500])
plt.title("Scatter Plot: Number of Letters in URL vs Ratio of Letters in URL")

plt.tight_layout()
plt.show()
```
